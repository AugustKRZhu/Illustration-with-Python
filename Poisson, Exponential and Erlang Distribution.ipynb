```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import exp, factorial
```

Show binomial with many n

```python
### use example of bus
## bus comes on average 1 buses in 4 hour
# lamda = rate of ouccurance = 1 in 4 hour
# Probability that will come in one hours
# ploting graph
problist = []
klist = []
n = 240
p = 0.25 # prob in 1 hour
q = 1 - p
for i in range(0,100):
    k = i
    klist.append(k)
    problist.append((factorial(n)/(factorial(k)*factorial(n-k)))*(p**k)*(q**(n-k)))
    
print("max probability",max(problist))
print("probability peak at k =",problist.index(max(problist)))
print("sum of all probability =",sum(problist))
print("np:", n*p)
plt.figure(figsize=(20,10))

plt.bar(klist,problist)
plt.plot(problist,color="red")


plt.show()
```

```python
### use example of bus
## bus comes on average 1 buses in 4 hour
# lamda = rate of ouccurance = 1 in 4 hour
# Probability that will come in one hours
# ploting graph
problist = []
klist = []
lamda = 0.25 # number of robbery in one hour
k = 5
for i in range(0,50):
    x = i
    klist.append(x)
    prob = 0
    for n in range(k):
        prob += (((lamda*x)**n)*exp((-lamda)*x))/factorial(n)
    problist.append(1-prob)
    
print("max probability",max(problist))
print("probability peak at k =",problist.index(max(problist)))
print("sum of all probability =",sum(problist))
plt.figure(figsize=(20,10))

plt.bar(klist,problist)
plt.plot(problist,color="red")


plt.show()
```
